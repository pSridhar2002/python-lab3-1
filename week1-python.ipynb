```python
x=set([1,3,4,2,2])
print(x)
x.update([4,5])
print(x)
y=set([2,4,5,9,8])
print(x.intersection(y))
print(x^y)
print(x.symmetric_difference(y))
print(x | y)
print(x & y)
print(x-y)
print(y-x)
print(x.remove(3))
x.clear()
print(x)
```
